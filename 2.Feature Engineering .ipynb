```python
# function to calculate CCI 
def calculateCCI(data):
    CCI = list() 
    score = 0 
    for i, j in data.iterrows(): 
        if( (j.Age >= 60) & (j.Age <= 69)): 
            score += 2
        if((j.Age >= 70) & (j.Age <= 79)): 
            score += 3
        if(j.Age >= 80): 
            score += 4
        comorbid_conditions = list(j.Chronic_Comorbid_Conditions.split(',')) 
        for ccc in comorbid_conditions: 
            if(ccc in cc1): 
                score += 1 
            else: 
                score += 2
        CCI.append(score)
        score = 0
    return CCI 
```

```python
CCI = calculateCCI(data) 
data = data.assign(CCI = pd.Series(CCI).values) 
```

```python
# function to calculate CPS (Comorbidity Polypharmacy Score)
def calculateCPS(data): 
    CPS = list() 
    for i, j in data.iterrows(): 
        comorbid_conditions = list(j.Chronic_Comorbid_Conditions.split(',')) 
        daily_medicine_intake = j.No_Of_Medicine_Per_Day
        # CPS = Comorbid conditions + Number of medication taken per day
        CPS.append(len(comorbid_conditions) + daily_medicine_intake) 
    return CPS 
```

```python
data = data.assign(CPS = pd.Series(calculateCPS(data)).values) 
data.shape 
```

```python
# grouping based on 'Symptom Burden' 
def symptom_burden(data): 
    burden = list() 
    for i, j in data.iterrows(): 
        prior_symptoms = len(j.Symptoms.split(','))
        post_medication_symptoms = len(j.Symptoms2.split('|')) 
        if(post_medication_symptoms < prior_symptoms): 
            burden.append('Lower') 
        else: 
            burden.append('Higher') 
    return burden
```

```python
data = data.assign(SYMPTOM_BURDEN = pd.Series(symptom_burden(data)).values)  
```

```python
# ten year survival Calculation 
def ten_year_survival_percentage(data):
    TEN_YEAR_SURVIVAL = list()
    for i, j in data.iterrows():
        SURVIVAL_PERCENTAGE = round(np.power(0.983, np.exp(j.CCI * 0.9)) * 100, 2)  
        TEN_YEAR_SURVIVAL.append(SURVIVAL_PERCENTAGE) 
    return TEN_YEAR_SURVIVAL
```

```python
TEN_YEAR_SURVIVAL_PERCENTAGE = ten_year_survival_percentage(data) 
data = data.assign(TEN_YEAR_SURVIVAL_PERCENTAGE = pd.Series(TEN_YEAR_SURVIVAL_PERCENTAGE).values) 
data.shape
```
